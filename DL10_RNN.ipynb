```python
docs = ['go bangladesh',
		'bangladesh bangladesh',
		'hip hip hurray',
		'joy bangla bangladesh jindabad',
		'amr shonar bangla',
		'shium shium',
		'i am a math graduate',
		'antu pappu naimur',
		'the name of our country is bangladesh',
		'inquilab zindabad']
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
tokenizer = Tokenizer()
```

```python
tokenizer.fit_on_texts(docs) # this will allocate every unique word a number
```

```python
tokenizer.word_index
```

```python
len(tokenizer.word_index) # number of unique words
```

```python
tokenizer.word_counts
```

```python
sequences = tokenizer.texts_to_sequences(docs)
sequences # So my sentence is now in the form of number sequence
```

```python
# Let's make every sequence of same size.

from keras.utils import pad_sequences
sequences = pad_sequences(sequences,padding='post')
sequences
```

## Let's try imdb dataset

```python
(X_train,y_train),(X_test,y_test) = imdb.load_data(num_words=10000)
```

```python
X_train
```

```python
X_train[0]
```

```python
len(X_train[0]) # this is the length of one review after tokenizing. Imdb data is already tokenized.
```

```python
print(X_train.shape)
print(X_test.shape)
```

```python
X_train = pad_sequences(X_train,padding='post',maxlen=500)
X_test = pad_sequences(X_test,padding='post',maxlen=500) # tokenized review is too long to train for my pc. So i took only 500 insted of 2494
```

```python
X_train.shape
```

```python
from tensorflow.keras.layers import Input

model = Sequential([
    Input(shape=(500,)),
    Embedding(input_dim=10000, output_dim=2),
    SimpleRNN(32),
    Dense(1, activation='sigmoid')
])

model.summary()
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['acc'])

history = model.fit(X_train, y_train,epochs=10,validation_data=(X_test,y_test))
```
